# ROZKŁAD NORMALNY

Możemy określić średnią i wariancję rozkładu normalnego za pomocą argumentów loc i scale do normy.rvs. Aby wygenerować 10000 liczb losowych ze średniej rozkładu normalnego = 0 i wariancji = 1, używamy funkcji norm.rvs jako

```python
from scipy.stats import norm
# generate random numbersfrom N(0,1)
data_normal = norm.rvs(size=10000,loc=0,scale=1)
ax = sns.distplot(data_normal,
                  bins=100,
                  kde=False,
                  color='red',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='normalna', ylabel='częstotliwość')
```

# ROZKŁAD Bernoulliego

```python
from scipy.stats import bernoulli
```

Zmienna losowa Bernoulliego może przyjąć wartość 0 lub 1, używając pewnego prawdopodobieństwa jako parametru. Aby wygenerować 10000 losowych liczb bernoulli z prawdopodobieństwem sukcesu p = 0,3, użyjemy bernoulli.rvs z dwoma argumentami.

```python
# generate bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.3)
ax= sns.distplot(data_bern,
                 kde=False,
                 color="skyblue",
                 hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Bernoulli', ylabel='Częstotliwość')
```

Z poniższego wykresu wynika, że na 10000 prób z prawdopodobieństwem powodzenia 0,3, osiągamy około 3000 sukcesów.

# ROZKŁAD POISONNA

```python
from scipy.stats import poisson
```

Wygenerujmy 10000 liczb losowych ze zmiennej losowej Poissona o mi = 0,3 i wykreślmy je.

```python
data_poisson = poisson.rvs(mu=3, size=10000)
ax = sns.distplot(data_poisson,
                  kde=False,
                  color='green',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Poisson', ylabel='Częstotliwość')
```

Wygenerujmy 10000 liczb losowych z rozkładu Beta z alfa = 5 i beta = 1. Histogram Beta (1,1) jest rozkładem jednolitym.

```python
data_beta = beta.rvs(5, 1, size=10000)
ax = sns.distplot(data_beta,
                  kde=False,
                  bins=100,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Beta(5,1)', ylabel='Częstotliwość')
```
